```python
import requests
from skimage import io
import matplotlib.pyplot as plt
```

```python
base_url = "http://localhost:8000/"
```

```python
file_path = "../samples/people_002.png"
file_info = {
    "file": open(file_path, 'rb'),
}
resp = requests.post(f"{base_url}predict", files=file_info)

img = io.imread(f"{base_url}{resp.json()['file_name']}")

plt.imshow(img)
```
